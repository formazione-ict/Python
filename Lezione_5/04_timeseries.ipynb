# ⏱️ Lezione 4: Time Series
Pandas è nato per la finanza: la gestione delle date è eccellente.

```python
import pandas as pd
import numpy as np

# Creiamo date giornaliere per un anno
date = pd.date_range(start='2024-01-01', periods=365, freq='D')
valori = np.random.randint(10, 100, 365)

ts = pd.DataFrame({'Data': date, 'Visite': valori})
ts.set_index('Data', inplace=True)
ts.head()
```

```python
# 1. Resampling (Cambio di frequenza)
# Calcoliamo la media MENSILE ('M') delle visite
mensile = ts.resample('M').mean()
mensile.head()
```

```python
# 2. Rolling Windows (Medie Mobili)
# Calcoliamo la media mobile a 7 giorni (smussa il grafico)
ts['Media_Mobile_7gg'] = ts['Visite'].rolling(window=7).mean()

# Plot veloce per vedere l'effetto
ts.iloc[:50].plot(figsize=(10, 5))
```
