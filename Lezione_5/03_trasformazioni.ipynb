Pivot Table per riassumere i dati e `.apply()` per logiche personalizzate.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Data': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Categoria': ['Elettronica', 'Abbigliamento', 'Elettronica', 'Abbigliamento'],
    'Vendite': [1000, 500, 1200, 600],
    'Profitto': [200, 150, 250, 180]
})
df
```

```python
# 1. Pivot Table
# Creiamo una matrice: Righe=Data, Colonne=Categoria, Valori=Somma Vendite
pivot = df.pivot_table(index='Data', 
                       columns='Categoria', 
                       values='Vendite', 
                       aggfunc='sum')
pivot
```

```python
# 2. Apply (Funzioni custom)
# Creiamo una funzione che classifica le vendite
def classifica_vendita(valore):
    if valore > 800:
        return 'Alta'
    return 'Bassa'

# Applichiamo la funzione riga per riga
df['Rating'] = df['Vendite'].apply(classifica_vendita)
df
```
