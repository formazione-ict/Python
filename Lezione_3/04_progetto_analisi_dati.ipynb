### 1. Creazione del Dataset
Normalmente caricheresti un CSV con `pd.read_csv('file.csv')`, ma qui lo creiamo al volo.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creiamo un dizionario di dati
dati = {
    'Studente': ['Mario', 'Luigi', 'Peach', 'Toad', 'Yoshi', 'Bowser', 'Wario'],
    'Matematica': [7, 8, 9, 6, 8, 4, 5],
    'Italiano': [6, 7, 9, 7, 6, 5, 4],
    'Informatica': [9, 8, 7, 9, 5, 6, 8]
}

# Trasformiamo in DataFrame (tabella)
df = pd.DataFrame(dati)

# Mostriamo le prime 5 righe
df.head()
```

### 2. Analisi Statistica
Pandas ci regala statistiche immediate.

```python
# Statistiche descrittive (media, min, max, quartili)
df.describe()
```

### 3. Calcolo Nuove Colonne
Calcoliamo la media di ogni studente.

```python
# Creiamo colonna 'Media_Totale'
df['Media_Totale'] = (df['Matematica'] + df['Italiano'] + df['Informatica']) / 3

# Arrotondiamo a 2 decimali
df['Media_Totale'] = df['Media_Totale'].round(2)

# Ordiniamo per i pi√π bravi
df_ordinato = df.sort_values(by='Media_Totale', ascending=False)
df_ordinato
```

### 4. Visualizzazione Finale

```python
plt.figure(figsize=(12, 6))

# Grafico a barre orizzontali
plt.barh(df_ordinato['Studente'], df_ordinato['Media_Totale'], color='skyblue')

plt.xlabel('Voto Medio')
plt.title('Classifica Studenti')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Linea della sufficienza
plt.axvline(x=6, color='red', linestyle='--', label='Sufficienza')
plt.legend()

plt.show()
```
