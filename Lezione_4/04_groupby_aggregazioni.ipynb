# ðŸ“Š Lezione 4: Aggregazione Dati (Group By)
Simile alle tabelle pivot di Excel. Raggruppa i dati per categoria e calcola statistiche.

```python
import pandas as pd

df = pd.DataFrame({
    'Dipartimento': ['IT', 'HR', 'IT', 'Sales', 'Sales', 'HR'],
    'Impiegato': ['Mario', 'Anna', 'Luigi', 'Sara', 'Paolo', 'Elena'],
    'Stipendio': [3000, 2000, 3200, 2500, 2800, 2100],
    'Esperienza_Anni': [5, 2, 7, 4, 5, 3]
})
df
```

```python
# 1. Group By Semplice: Stipendio medio per dipartimento
media_dip = df.groupby('Dipartimento')['Stipendio'].mean()
print(media_dip)
```

```python
# 2. Aggregazioni Multiple (.agg)
# Per ogni dipartimento vogliamo: stipendio medio, massimo e numero impiegati
report = df.groupby('Dipartimento')['Stipendio'].agg(['mean', 'max', 'count'])

# Rinominiamo le colonne per chiarezza
report.columns = ['Media', 'Max', 'Numero_Staff']
report
```
