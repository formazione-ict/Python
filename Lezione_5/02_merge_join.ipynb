Come fare i VLOOKUP (Cerca.Vert) o le JOIN di SQL in Pandas.

```python
import pandas as pd

# Tabella Clienti
clienti = pd.DataFrame({
    'ID_Cliente': [1, 2, 3, 4],
    'Nome': ['Alice', 'Bob', 'Charlie', 'David'],
    'CittÃ ': ['Roma', 'Milano', 'Torino', 'Napoli']
})

# Tabella Ordini (David non ha ordini, e c'Ã¨ un ordine di ID 99 sconosciuto)
ordini = pd.DataFrame({
    'ID_Ordine': [101, 102, 103, 104],
    'ID_Cliente': [1, 2, 1, 99],
    'Importo': [250, 100, 50, 500]
})
print("Clienti:")
display(clienti)
print("Ordini:")
display(ordini)
```

```python
# 1. INNER JOIN (Default)
# Tiene solo le righe che hanno corrispondenza in ENTRAMBE le tabelle
# David e l'ordine 99 spariscono
inner = pd.merge(clienti, ordini, on='ID_Cliente', how='inner')
inner
```

```python
# 2. LEFT JOIN
# Tiene TUTTI i clienti. Se non hanno ordini, mette NaN (vedi David)
left = pd.merge(clienti, ordini, on='ID_Cliente', how='left')
left
```

```python
# 3. OUTER JOIN
# Tiene TUTTO da entrambe le parti. 
outer = pd.merge(clienti, ordini, on='ID_Cliente', how='outer')
outer
```
