Analizzeremo un dataset di vendite per rispondere a domande di business.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Creazione dataset finto (per rendere l'esempio eseguibile ovunque)
data = {
    'Data': pd.date_range(start='2024-01-01', periods=100),
    'Prodotto': np.random.choice(['Smartphone', 'Laptop', 'Tablet', 'Cuffie'], 100),
    'Regione': np.random.choice(['Nord', 'Centro', 'Sud'], 100),
    'Prezzo': np.random.uniform(50, 1000, 100).round(2),
    'Qta': np.random.randint(1, 10, 100)
}
df = pd.DataFrame(data)

# Calcoliamo il fatturato per riga
df['Fatturato'] = df['Prezzo'] * df['Qta']
df.head()
```

```python
# 2. Analisi: Qual √® il prodotto pi√π venduto?
top_prodotti = df.groupby('Prodotto')['Fatturato'].sum().sort_values(ascending=False)
top_prodotti.plot(kind='bar', color='teal', title='Fatturato per Prodotto')
plt.ylabel('Euro')
```

```python
# 3. Analisi Temporale: Vendite per Mese
# Estraiamo il mese dalla data
df['Mese'] = df['Data'].dt.month_name()

trend = df.groupby('Mese')['Fatturato'].sum()
trend
```
