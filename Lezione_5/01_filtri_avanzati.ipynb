Andiamo oltre il semplice `df[df['col'] > 5]`. 
Esploriamo metodi piÃ¹ potenti e leggibili come `.query()` e `.isin()`.

```python
import pandas as pd

df = pd.DataFrame({
    'Nome': ['Mario', 'Luigi', 'Peach', 'Toad', 'Wario'],
    'Ruolo': ['Idraulico', 'Idraulico', 'Principessa', 'Assistente', 'Cattivo'],
    'Monete': [150, 80, 500, 20, 1000],
    'Regno': ['Funghi', 'Funghi', 'Funghi', 'Funghi', 'Sabbia']
})
df
```

```python
# 1. Metodo .isin() - Alternativa a molti OR
# Invece di: (df['Ruolo'] == 'Idraulico') | (df['Ruolo'] == 'Principessa')
eroi = df[df['Ruolo'].isin(['Idraulico', 'Principessa'])]
eroi
```

```python
# 2. Metodo .query() - Sintassi simil-SQL (molto leggibile)
soglia = 100
# Nota come possiamo usare variabili esterne con @
ricchi = df.query("Monete > @soglia and Regno == 'Funghi'")
ricchi
```

```python
# 3. Filtri su Stringhe (.str)
# Trova chi ha la lettera 'a' nel nome (case insensitive)
con_a = df[df['Nome'].str.contains('a', case=False)]

# Negazione con ~ (Tilde)
non_idraulici = df[~df['Ruolo'].str.contains('Idraulico')]
non_idraulici
```
