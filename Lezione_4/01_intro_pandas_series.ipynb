Pandas √® la libreria standard per la manipolazione dati in Python. 
Le due strutture principali sono:
1. **Series**: array 1D etichettato (una colonna).
2. **DataFrame**: tabella 2D (come un foglio Excel).

```python
import pandas as pd
import numpy as np

# 1. Creazione di una SERIES
voti = [28, 30, 25, 18, 24]
studenti = ['Mario', 'Luigi', 'Peach', 'Toad', 'Yoshi']

s = pd.Series(data=voti, index=studenti)
print("--- La mia prima Series ---")
print(s)
```

```python
# 2. Accesso ai dati nella Series
print(f"Voto di Peach: {s['Peach']}")
print(f"Voti sufficienti (filtro):\n{s[s >= 18]}")
```

```python
# 3. Creazione DataFrame da Dizionario
data = {
    'Nome': studenti,
    'Voto': voti,
    'Citt√†': ['Roma', 'Milano', 'Napoli', 'Torino', 'Bari']
}

df = pd.DataFrame(data)
df.head() # Mostra le prime 5 righe
```
