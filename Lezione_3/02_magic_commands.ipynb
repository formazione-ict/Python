### 3. Magic per l'intera cella (`%%`)
Esempio: `%%time` misura quanto impiega *tutta la cella* ad eseguirsi (una sola volta, non media).

```python
%%time

# Simuliamo un calcolo pesante
somma = 0
for i in range(1000000):
    somma += i
    
print(f"Finito! Somma: {somma}")
```
