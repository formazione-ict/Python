Impareremo la differenza cruciale tra `loc` (per etichette) e `iloc` (per indici numerici).

```python
import pandas as pd

# Creiamo un DataFrame di esempio
df = pd.DataFrame({
    'Prodotto': ['Laptop', 'Mouse', 'Monitor', 'Tastiera', 'Cavo'],
    'Prezzo': [1200, 25, 300, 50, 10],
    'Qta': [5, 100, 20, 50, 200]
})
# Impostiamo 'Prodotto' come indice (etichetta delle righe)
df.set_index('Prodotto', inplace=True)
df
```

```python
# 1. Uso di .loc (Label based)
# Voglio il prezzo del Monitor
print(f"Prezzo Monitor: {df.loc['Monitor', 'Prezzo']} â‚¬")

# 2. Uso di .iloc (Integer based)
# Voglio la prima riga (Laptop) e la seconda colonna (Qta)
print(f"Qta primo prodotto: {df.iloc[0, 1]}")
```

```python
# 3. Filtraggio Avanzato (Maschere Booleane)
# Prodotti costosi (> 100â‚¬) E con scorte basse (< 10)
critici = df[(df['Prezzo'] > 100) & (df['Qta'] < 10)]
critici
```
