I dati reali sono sporchi. Vediamo come gestire valori mancanti (`NaN`) e creare nuove colonne.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Cliente': ['A', 'B', 'C', 'D'],
    'Ordini': [10, np.nan, 5, 2],  # np.nan simula un valore mancante
    'Spesa': [1000, 500, np.nan, 200]
})
df
```

```python
# 1. Gestione Valori Mancanti
# Riempiamo i NaN negli ordini con 0
df['Ordini'] = df['Ordini'].fillna(0)

# Riempiamo la spesa mancante con la media
media_spesa = df['Spesa'].mean()
df['Spesa'] = df['Spesa'].fillna(media_spesa)
df
```

```python
# 2. Creazione Nuove Colonne (Feature Engineering)
df['Spesa_Media_Ordine'] = df['Spesa'] / df['Ordini']

# Attenzione: divisione per zero genera 'inf' (infinito)
# Sostituiamo inf con 0
df.replace([np.inf, -np.inf], 0, inplace=True)
df
```
